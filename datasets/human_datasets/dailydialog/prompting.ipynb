```python
from transformers import AutoTokenizer
import transformers
import torch
import pandas as pd
```

```python
from transformers import AutoModelForCausalLM

model_id="tiiuae/falcon-7b-instruct"
tokenizer=AutoTokenizer.from_pretrained(model_id)
model=AutoModelForCausalLM.from_pretrained(model_id, trust_remote_code=True)
```

```python
data = pd.read_json('data.ndjson', lines=True)
```

```python
def process_example(e):
    inputs = []
    for i, t in enumerate(e.split(' [EOT] ')):
        if i%2==0:
            inputs.append(f'person 1: {t}')
        else:
            inputs.append(f'person 2: {t}')
    if len(inputs) % 2 == 0:
        inputs.append('person 1: ')
    else:
        inputs.append('person 2: ')
    return ' '.join(inputs)
```

```python
inputs = [process_example(e) for e in data['source'].tolist()]
```

```python
inputs
```

```python
encoded_ids = [tokenizer(i, return_tensors="pt", return_token_type_ids=False) for i in inputs]
```

```python
ls = []
for i in encoded_ids:
    ls.append(i['input_ids'].shape[-1])
```

```python
outputs = model.generate(**encoded_ids[0], min_length=min(ls), max_length=max(ls))
```
